```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# read in the data

movies_df = pd.read_csv('/Users/farihahazad/Downloads/movies.csv')
```

```python
# look at the data

movies_df
```

```python
# check for missing data
# loop through the data to figure out what's missing

for col in movies_df.columns:
    pct_missing = np.mean(movies_df[col].isnull())
    print ('{} - {}%'.format(col, round(pct_missing*100)))
    
```

```python
# drop rows with missing data

movies_df = movies_df.dropna()
movies_df
```

```python
# check again for missing data 

for col in movies_df.columns:
    pct_missing = np.mean(movies_df[col].isnull())
    print ('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# data types for our columns 

movies_df.dtypes
```

```python
# change data type of columns

movies_df['budget'] = movies_df['budget'].astype('int64')

movies_df['gross'] = movies_df['gross'].astype('int64')

```

```python
# create correct year column

movies_df['year_correct'] = movies_df['released'].str.extract(pat='([0-9]{4})').astype(int)
```

```python
movies_df
```

```python
# order the data

movies_df = movies_df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# drop duplicates

movies_df.drop_duplicates()
```

```python
# scatter plot with budget vs gross

plt.scatter(x=movies_df['budget'],y=movies_df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
movies_df.head()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=movies_df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# look at correlation between all numeric columns

movies_df.corr(method= 'pearson')
```

```python
movies_df.corr()
```

```python
# there is a high correlation between budget and gross
```

```python
correlation_matrix = movies_df.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Metric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# assign random numeric value for each unique catergorical value

movies_df_numerized = movies_df

for col_name in movies_df_numerized.columns:
    if (movies_df_numerized[col_name].dtype == 'object'):
        movies_df_numerized[col_name] = movies_df_numerized[col_name].astype('category')
        movies_df_numerized[col_name] = movies_df_numerized[col_name].cat.codes
        
movies_df_numerized
```

```python
correlation_matrix = movies_df_numerized.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Metric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
movies_df_numerized.corr()
```

```python
correlation_mat = movies_df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# look at the ones with a high correlation (>0.5)

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
